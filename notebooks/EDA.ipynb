```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ts = pd.read_csv('../datasets/vic_elec.csv')
ts
```

```python
ts.info()
```

```python
ts['Time'] = pd.to_datetime(ts['Time'])
ts = ts.set_index('Time')
```

```python
fig, ax = plt.subplots(figsize=(18, 4))
plt.title('Demanda ElÃ©ctrica')
sns.lineplot(x=ts.index, y=ts.Demand, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ts['anio'] = ts.index.year
ts.boxplot(column='Demand', by='anio', ax=ax,)
ts.groupby('anio')['Demand'].median().plot(style='o-', linewidth=0.8, ax=ax)
ax.set_ylabel('Demand')
ax.set_title('DistribuciÃ³n demanda por mes')
fig.suptitle('');
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ts['mes'] = ts.index.month
ts.boxplot(column='Demand', by='mes', ax=ax,)
ts.groupby('mes')['Demand'].median().plot(style='o-', linewidth=0.8, ax=ax)
ax.set_ylabel('Demand')
ax.set_title('DistribuciÃ³n demanda por mes')
fig.suptitle('');
```
